```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import re
import joblib
import string
```

```python
fake = pd.read_csv('C:/Users/venka/Desktop/FakeNews/Fake.csv')
true = pd.read_csv('C:/Users/venka/Desktop/FakeNews/True.csv')
```

```python
fake.head()
```

```python
true.head()
```

```python
fake['class'] = 0
true['class']  = 1
```

```python
data = pd.concat([fake,true],axis=0)
```

```python
data.sample(10)
```

```python
data = data.drop(["title","subject","date"],axis=1)
```

```python
data.sample(10)
```

```python
data.reset_index(inplace=True)
```

```python
data.drop(["index"],inplace=True,axis=1)
```

```python
data.sample(5)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
data["text"] = data["text"].apply(clean_text)
```

```python
data.sample(5)
```

```python
x=data["text"]
y=data["class"]

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
xv_train = vectorizer.fit_transform(xtrain)
xv_test = vectorizer.transform(xtest)
```

```python
model = LogisticRegression()
model.fit(xv_train, ytrain)
```

```python
prediction = model.predict(xv_test)
model.score(xv_test,ytest)
```

```python
print(classification_report(ytest,prediction))
```

```python
joblib.dump(vectorizer,'vectorizer.jb')
joblib.dump(model,'model_lr.jb')
```
